```python
#slicing

#substring[start:stop:step]

text='India is Great'
print(text[0:5])
print(text[6:9])
print(text[9:14])
print(text[-14:-9])

print("----------------------------------------")


print(text[-9::-1])
print(text[0:8:3])



print("----------------------------------------")
#reverse using positive index
print(text[13:8:-1])



print("----------------------------------------")
#extracting of iii
print(text[0:7:3])


print("--------------------------------------")




```
