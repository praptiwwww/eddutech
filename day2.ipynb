```python
my_info = {'name':'Alice', 'age':25, 'City':'Newyork'}

print(my_info, type(my_info))
# 1. clear() – Removes all items from the dictionary

my_info.clear()
print(my_info)

my_info = {'name':'Alice', 'age':25, 'City':'Newyork'}

# 2. copy() – Returns a shallow copy of the dictionary

copy_info = my_info.copy()
print(copy_info)

# 3. fromkeys() – Creates a new dictionary with specified keys and a common value

keys = ['name', 'age', 'city']
default_value = None

new_dict = dict.fromkeys(keys, default_value)
print(new_dict)

# 4. get() – Returns the value for a specified key, or None if the key is not found

print(my_info.get('name', 'Not Found'))
print(my_info.get('gender', 'Not Found'))

# 5. items() – Returns a view object of key-value pairs

print(my_info.items())

print(list(my_info.items()))

# 6. keys() – Returns a view object of dictionary keys

print(my_info.keys())

print(list(my_info.keys()))

# 7. pop() – Removes and returns the value for a specified key

my_info = {'name':'Alice', 'age':25, 'City':'Newyork'}

age = my_info.pop('age')

print(age)
print(my_info)

# 8. popitem() – Removes and returns the last key-value pair

last_item = my_info.popitem()
print(last_item)
print(my_info)


# 9. setdefault() – Returns the value of a key, or sets it to a default if not found

name = my_info.setdefault('name', 'Unknown')
print(name)

gender = my_info.setdefault('gender', 'Female')
print(gender)
print(my_info)
```

```python
a = int(input("enter marks 1:"))
b = int(input("enter marks 1:"))
c = int(input("enter marks 1:"))

avg = (a + b + c) / 3
print('Average marks obtain={avg}')

if (avg >= 91) and (avg <= 100):
  print('A+')
elif (avg >= 81) and (avg <= 90):
  print('A')
 elif (avg >= 71) and (avg <= 80):
  print('B+')
elif (avg >= 61) and (avg <= 70):
  print('B')
elif (avg >= 51) and (avg <= 60):
  print('C')
elif (avg >= 41) and (avg <= 50):
  print('D')
else:
  print('F')
```
