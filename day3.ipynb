```python
#arbitary element
#using of infinite elements

def add_number(*num):
  result=0
  for i in num:
    result-=i
    return result
add_number(5,8,-10,6,5)


```

```python
#multiplication
def mul(*numm):
  result=1
  for i in numm:
    result*=i
    return result
mul(8,6,4,3)
```

```python
## double star arguments
def car(**detail):
  for k,v in details.item():
    print(f'{k}:{v}')

  car("brand"="toyota":"model"="30c")
  print(detail)
```

```python
def find_max(*val):
  return max(val)

find_max(3,5,6)
```

```python
#lamda function
#all function in one line

num=50
if num%2==0:
  print("even")
else:
  print("odd")

odd_even=lambda num:"even" if num%2==0 else"odd"
odd_even(50)

```

```python
sq=lambda side:side*side if side%2==0 else side*side*side
sq(2)

```

```python
#a,b,c which one is the largest------>

largest = lambda a, b, c: a if ((a>=b) and (a>=c)) else (b if ((b>=a) and (b>=c)) else c)

largest(40, 30, 30)
```
